### Plotting univariate graphs for each variable

```python
# Histogram
plt.rcParams["figure.figsize"] = (9, 7)
plt.hist(gapminder_df['incomeperperson'])
plt.xlabel("Income per person", fontsize=15)
plt.title("Income per person - Worldwide", fontsize=20, color='green')
plt.show()
```

```python
plt.rcParams["figure.figsize"] = (9, 7)
plt.hist(gapminder_df['suicideper100th'])
plt.xlabel("Suicide per 100TH", fontsize=15)
plt.title("Suicide per Hunderd Thousand - Worldwide", fontsize=20, color='green')
plt.show()
```

```python
plt.rcParams["figure.figsize"] = (9, 7)
plt.hist(gapminder_df['alcconsumption'])
plt.xlabel("Alcohol Consumption", fontsize=15)
plt.title("Alcohol Consumption - Worldwide", fontsize=20, color='green')
plt.show()
```

### Plotting bivariate Scatter plot for quantitative values.

```python
plt.scatter(gapminder_df['incomeperperson'], gapminder_df['suicideper100th'])
plt.rcParams["figure.figsize"] = (9, 7)
plt.title("Income per Person Vs Suicide rate", fontsize=20, color='green')
plt.xlabel("Income per Person", fontsize=15)
plt.ylabel("Suicide per 100th", fontsize=15)

m, b = np.polyfit(gapminder_df['incomeperperson'],gapminder_df['suicideper100th'], 1)
plt.plot(gapminder_df['incomeperperson'], m*gapminder_df['incomeperperson'] + b)

plt.show()
```

```python
plt.scatter(gapminder_df['incomeperperson'], gapminder_df['alcconsumption'])
plt.rcParams["figure.figsize"] = (9, 7)
plt.title("Income per Person Vs Alcohol Consumption", fontsize=20, color='green')
plt.xlabel("Income per Person", fontsize=15)
plt.ylabel("Alcohol Consumption", fontsize=15)

m, b = np.polyfit(gapminder_df['incomeperperson'],gapminder_df['alcconsumption'], 1)
plt.plot(gapminder_df['incomeperperson'], m*gapminder_df['incomeperperson'] + b)

plt.show()
```
